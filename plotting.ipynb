```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

plt.rcParams['font.size'] = 18

# Importing
df = pd.read_csv('Generationlength.csv')

# Means
means = []
for i in df.columns[1:]:
    means.append(np.mean(df[i]))
# Plotting
fig, ax = plt.subplots(figsize = (18, 10))
ax.set_ylabel('Time')
sns.boxplot(data = df, color = 'cornflowerblue', ax = ax)
sns.swarmplot(data=df, palette='dark:k', size = 6, ax = ax)
sns.scatterplot()
plt.legend()
plt.show()
```
